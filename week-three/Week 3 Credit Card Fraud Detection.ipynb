```python
#Load the datasets from kaggle (the Jupyter Notebook has been created and run on the Kaggle website)

train_identity = pd.read_csv(f'/kaggle/input/ieee-fraud-detection/train_identity.csv')
train_transaction = pd.read_csv(f'/kaggle/input/ieee-fraud-detection/train_transaction.csv')
test_identity = pd.read_csv(f'/kaggle/input/ieee-fraud-detection/test_identity.csv')
test_transaction = pd.read_csv(f'/kaggle/input/ieee-fraud-detection/test_transaction.csv')
sub = pd.read_csv(f'/kaggle/input/ieee-fraud-detection/sample_submission.csv')

# let's combine the data and work with the whole dataset
train = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')
test = pd.merge(test_transaction, test_identity, on='TransactionID', how='left')
```

```python
full_train_df = pd.merge(train_identity,train_transaction, on='TransactionID', how='left')
```

Lots of columns, and also there seems to be a lot of NaN data, that needs to be adressed before training.

**Visualisation of records without fraud vs fraudulent records**

```python
import seaborn as sns
sns.countplot(x='isFraud', data=full_train_df)
```

OK, looks like there is an overwhelming number of valid transactions and only a tiny fraction of fraudulent transactions. So we will most likely get a high success rate after training, which will, in fact, be misleading. 

We also see some  NaN  data in the previews above, so let's fill them with something useful.

```python
full_train_df.fillna(value=full_train_df.median(), inplace=True)
```

Also, we got some non-numerical colums, which need to be translated before training.

```python
from sklearn  import preprocessing
for col in full_train_df.columns:
    if full_train_df[col].dtype=='object' :
        print("label encoding",col)
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(full_train_df[col].values))
        full_train_df[col] =lbl.transform(list(full_train_df[col].values))
```

Now lets check if there are any correlations of the data somehow. I love heatmaps, so let's use them:

```python
# Correlation matrix
corrmat = full_train_df.corr()
sns.heatmap(corrmat, square = False)
```

Good, a lot of red and yellow areas, but also the white areas look like they got no useful data at all. To get rid of them, I am playing with several thresholds of when a set of data should be dropped. Lets start with 40% empty data, and look at the result in another heatmap.

```python
df_clean = full_train_df.loc[:, full_train_df.isnull().median() <= .4]

# Another correlation matrix
corrmat = df_clean.corr()
sns.heatmap(corrmat, square = False)
```

Nice, almost all of the white space is gone. Can we even keep more? Trying with 80%

```python
df_clean = full_train_df.loc[:, full_train_df.isnull().median() <= .8]
# Another correlation matrix
corrmat = df_clean.corr()
sns.heatmap(corrmat, square = False)
```

OK, cannot spot any difference, so we will leave it at that. Now lets train the data, with a train/test ratio of 1:1 (50% training, 50% test data)

```python
x_clean_df = df_clean.drop(["isFraud"], axis=1)
y_clean_df = df_clean["isFraud"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x_clean_df, y_clean_df, test_size=0.50)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
```

```python
logReg = LogisticRegression(solver="lbfgs", max_iter=4000)
logReg.fit(X_train, y_train)
```

Training complete, let's test and see, what we got.

```python
logReg.score(X_test, y_test)
```

0.92 is actually not that good, given that we only had a tiny fraction of fraudulent actions. Let's make a prediction and check this out.

```python
y_pred=logReg.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
# Confusion Matrix (by https://blog.usejournal.com/credit-card-fraud-detection-by-neural-network-in-keras-4bd81cc9e7fe)
cm = confusion_matrix(y_test, y_pred) # rows = truth, cols = prediction
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True, fmt='g')
print("Test Data Accuracy: %0.4f" % accuracy_score(y_test, y_pred))
```

Not good, the prediction failed to classify most of the fraudulend cases.
